### Ratings

Each integer rating should be interpreted as follows:

| Rating      | Meaning         |
| ----------- | --------------- |
| -5          | Hated it        |
| -3          | Didn't like it  |
| 0           | Haven't read it |
| 1           | OK              |
| 3           | Liked it        |
| 5           | Really liked it |

### The similarity algorithm

If user `a` enjoyed reading a number of books that user `b` also enjoyed reading, then when user `a` recommends a book that user `b` has not read, it is likely that user `b` would also enjoy reading it. On the other hand, if the two users tend to disagree about books, then they are unlikely to read each other's recommendations.

You can calculate how similar two users are by treating each of their list of book ratings as a vector, and then by calculating the dot product of these two vectors. The dot product is simply the sum of the products of each of the corresponding elements.

For example, suppose we have 3 books in our data file, and:
- user `a` has rated them [5, 3, -5] (here, user `a` has given books 1, 2, and 3 the ratings 5, 3, and -5 respectively);
- user `b` has rated them [1, 5, -3];
- user `c` has rated them [5, -3, 5]; and
- user `d` has rated them [1, 3, 0].

The similarity between users `a` and `b` is calculated as:

$$similarity(a, b) = \vec{ratings_a} \cdot \vec{ratings_b} = (5 \times 1) + (3 \times 5) + (-5 \times -3) = 35$$

Next, the similarity between user `a` and user `c` is:

$$similarity(a, c) = \vec{ratings_a} \cdot \vec{ratings_c} = (5 \times 5) + (3 \times -3) + (-5 \times 5) = -9$$

Finally, the similarity between user `a` and user `d` is:

$$similarity(a, d) = \vec{ratings_a} \cdot \vec{ratings_d} = (5 \times 1) + (3 \times 3) + (-5 \times 0) = 14$$

From this algorithm, we can see that if two people like a book (that is, they both rate it with a positive number) or dislike a book (that is, they both rate it negatively), then their similarity increases. If their ratings differ (that is, one gives it a positive rating, and the other a negative rating), then their similarity will decrease. The similarity score is unchanged where one or other of the users hasn't read the book.

Once you have calculated the pair-wise similarity between user `a` and every other user, you can then identify whose ratings are most similar to user `a`'s. In the above example, user `a` is most similar to user `b`, so we would recommend to user `a` the top-rated books from user `b` that user `a` has not already read.
